```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
lasso = Lasso(alpha=0.000002, copy_X=True, fit_intercept=True,random_state=3, max_iter=1000, tol=0.0001,
              precompute=False, warm_start=1, positive=False, selection='random')
```

```python
lasso.fit(X, y)
```

```python
selected_features = np.arange(1463)[lasso.coef_ !=0]
print("Selected Features: ", selected_features)
```

```python
list=[]
count=0
for i in selected_features:
    #print(df.columns[i+1])
    list.append(df.columns[i+1])
    count+=1
```
