```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV,train_test_split
from sklearn.datasets import load_diabetes
```

```python
#Protein_final_utf_8.csv
```

```python
dataset = pd.read_csv('Protein_final_ut.csv',
                 lineterminator='\n')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
X=dataset.drop(columns='Class Label\r',axis=1)
y=dataset['Class Label\r']
```

```python
print(X)
```

```python
print(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
pipeline=Pipeline([
    ('scaler',StandardScaler()),
    ('model',Lasso())
])
```

```python
search=GridSearchCV(pipeline,
                   {'model__alpha':np.arange(0.1,3,0.1)},
                    cv=5,
                    scoring='neg_mean_squared_error',
                    verbose=3
                   )
```

```python
search.fit(X_train,y_train)
```

```python
search.best_params_
```

```python
coefficients = search.best_estimator_.named_steps['model'].coeff_
```
